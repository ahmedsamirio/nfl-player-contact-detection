```python
import glob
import subprocess

def video_with_pair_contact(pair_df, tr_helmets, gp, frames_path, verbose=True):
    # Create output video
    VIDEO_CODEC = "MP4V"
    
    fps = 59.94/2
    width = 1280*2
    height = 720
    
    output_path = f"contact_{gp}.mp4"
    tmp_output_path = "tmp_" + output_path.split('.')[0] + ".mp4"
    
    output_video = cv2.VideoWriter(
        tmp_output_path, cv2.VideoWriter_fourcc(*VIDEO_CODEC), fps, (width, height)
    )
    
    # Filter gameplay pairs
    sideline_pair_df = pair_df.query('game_play == @gp and view == "Sideline"')
    sideline_helmets_df = tr_helmets.query('game_play == @gp and view == "Sideline"')
    
    endzone_pair_df = pair_df.query('game_play == @gp and view == "Endzone"')
    endzone_helmets_df = tr_helmets.query('game_play == @gp and view == "Endzone"')
    
    # Glob gameplay frames
    sideline_frames_paths = sorted(glob.glob(frames_path+f'/{gp}_Sideline*'))
    endzone_frames_paths = sorted(glob.glob(frames_path+f'/{gp}_Endzone*'))
    
    # Loop over frames
    for frame_no, (sideline_fp, endzone_fp) in tqdm(enumerate(zip(sideline_frames_paths, endzone_frames_paths))):
        
        if frame_no < 289:
            continue
               
        final_frame = []
        
        for view in ['Sideline', 'Endzone']:
            
            if view == 'Sideline':
                
                 # Read frame
                frame = read_img(sideline_fp)
                # frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

                # Make copy for predictions
                frame_pred = frame.copy()
        
                # Filter frame pairs and helmets
                frame_pair_df = sideline_pair_df.query('frame == @frame_no and contact == 1')
                frame_pair_preds_df = sideline_pair_df.query('frame == @frame_no and contact_pred_rolling > thresh')
                frame_helmets_df = sideline_helmets_df.query('frame == @frame_no')
                
            else:
                
                # Read frame
                frame = read_img(endzone_fp)
                # frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

                # Make copy for predictions
                frame_pred = frame.copy()
                
                # Filter frame pairs and helmets
                frame_pair_df = endzone_pair_df.query('frame == @frame_no and contact == 1')
                frame_pair_preds_df = endzone_pair_df.query('frame == @frame_no and contact_pred_rolling > thresh')
                frame_helmets_df = endzone_helmets_df.query('frame == @frame_no')
                

            # Create set for plotted helmets
            players_set = set()

            # Plot helmets if exists
            if len(frame_helmets_df) > 0:

                # Loop over helmets
                for helmet_tuple in frame_helmets_df.itertuples():

                    # Plot helmet if wasn't plotted before
                    if helmet_tuple.player_label not in players_set:

                        frame = plot_helmet_on_frame(frame, helmet_tuple.left, helmet_tuple.width, helmet_tuple.top, helmet_tuple.height, color=(0, 0, 0))
                        frame = plot_player_name_on_frame(frame, helmet_tuple.player_label, helmet_tuple.left, helmet_tuple.top, color=(0, 0, 0))

                        frame_pred = plot_helmet_on_frame(frame_pred, helmet_tuple.left, helmet_tuple.width, helmet_tuple.top, helmet_tuple.height, color=(0, 0, 0))
                        frame_pred = plot_player_name_on_frame(frame_pred, helmet_tuple.player_label, helmet_tuple.left, helmet_tuple.top, color=(0, 0, 0))


            # Plot pred pairs if exits
            if len(frame_pair_preds_df) > 0:

                # Loop over pairs
                for contact_tuple in frame_pair_preds_df.itertuples():
                    
                    if contact_tuple.bbox_distance < 150:
                    
                        # Plot second player helmet and name
                        if not np.isnan(contact_tuple.left_2):
                            helm_color = (0, 255, 0)

                            frame_pred = plot_helmet_on_frame(frame_pred, contact_tuple.left_2, contact_tuple.width_2, contact_tuple.top_2, contact_tuple.height_2, color=helm_color)
                            frame_pred = plot_player_name_on_frame(frame_pred, contact_tuple.player_label_2, contact_tuple.left_2, contact_tuple.top_2, color=helm_color)

                            # Plot line between players
                            frame_pred = plot_contact_line(frame_pred, contact_tuple)

                            players_set.add(contact_tuple.player_label_2)

                        else:
                            helm_color = (255, 0, 0)

                        # Plot first player helmet and name
                        frame_pred = plot_helmet_on_frame(frame_pred, contact_tuple.left_1, contact_tuple.width_1, contact_tuple.top_1, contact_tuple.height_1, color=helm_color)
                        frame_pred = plot_player_name_on_frame(frame_pred, contact_tuple.player_label_1, contact_tuple.left_1, contact_tuple.top_1, color=helm_color)

                        # Add both players to plotted set
                        players_set.add(contact_tuple.player_label_1)

            final_frame.append(frame_pred)
        
        final_frame = np.concatenate(final_frame, axis=1)
        output_video.write(final_frame)
        
    output_video.release()
    
    # Not all browsers support the codec, we will re-load the file at tmp_output_path
    # and convert to a codec that is more broadly readable using ffmpeg
    if os.path.exists(output_path):
        os.remove(output_path)
        
    subprocess.run(
        [
            "ffmpeg",
            "-i",
            tmp_output_path,
            "-crf",
            "18",
            "-preset",
            "veryfast",
            "-hide_banner",
            "-loglevel",
            "error",
            "-vcodec",
            "libx264",
            output_path,
        ]
    )
    
    os.remove(tmp_output_path)

    return output_path

def read_img(path):
    frame = cv2.imread(path)
    return frame
```
